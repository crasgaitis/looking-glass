```python
import pandas as pd

df = pd.read_csv('test/eeg.csv')
df.head()
```

```python
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
import numpy as np

def apply_pca(X, min_explained_variance=0.80):

    pca = PCA()
    pca.fit(X)

    explained_variance_ratio_cumsum = np.cumsum(pca.explained_variance_ratio_)
    plt.plot(explained_variance_ratio_cumsum, marker='o')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.title('Cumulative Explained Variance Plot')

    min_components = np.argmax(explained_variance_ratio_cumsum >= min_explained_variance) + 1

    plt.axhline(min_explained_variance, color='r', linestyle='--')
    plt.axvline(min_components, color='r', linestyle='--')

    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    plt.show()

    print(f"Minimum components required for {min_explained_variance:.0%} explained variance: {min_components}")
    
    return pca

pca1 = apply_pca(df)
```
